# Ex - GroupBy


### Introduction:

GroupBy can be summarized as Split-Apply-Combine.

Special thanks to: https://github.com/justmarkham for sharing the dataset and materials.

### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv).

### Step 3. Assign it to a variable called drinks.

```python
drinks = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv')
drinks.head()
```

### Step 4. Which continent drinks more beer on average?

```python
drinks.groupby('continent').beer_servings.mean()
```

### Step 5. For each continent print the statistics for wine consumption.

```python
drinks.groupby('continent').wine_servings.describe()
```

### Step 6. Print the mean alcohol consumption per continent for every column

```python
drinks.groupby('continent').agg({'beer_servings': 'mean', 'wine_servings': 'mean',
                                 'spirit_servings': 'mean', 'total_litres_of_pure_alcohol': 'mean'})
# The above code selects only the numeric columns for calculating the mean within each continent group.
# This avoids trying to calculate the mean of the 'country' column, which is of object (string) type.
```

### Step 7. Print the median alcohol consumption per continent for every column

```python
drinks.groupby('continent').agg({'beer_servings': 'median', 'wine_servings': 'median',
                                 'spirit_servings': 'median', 'total_litres_of_pure_alcohol': 'median'})
# Instead of applying median to the entire DataFrame after groupby, we specifically
# select the numeric columns we want to calculate the median for.
# This avoids trying to calculate the median of the 'country' column, which is of object (string) type.
```

### Step 8. Print the mean, min and max values for spirit consumption.
#### This time output a DataFrame

```python
drinks.groupby('continent').spirit_servings.agg(['mean', 'min', 'max'])
```
